```python
import numpy as np

friends = ["Alice", "Bob", "Carol","Danial"]

n = len(friends)
expense_matrix = np.zeros((n, n))

def add_expense(payer, beneficiaries, amount):
    payer_idx = friends.index(payer)
    split_amount = amount / len(beneficiaries)

    for person in beneficiaries:
        person_idx = friends.index(person)
        expense_matrix[payer_idx][person_idx] += split_amount

add_expense("Alice", ["Alice", "Bob", "Carol","Danial"], 600)
add_expense("Bob", ["Bob", "Carol","Danial"], 200)
add_expense("Carol", ["Alice", "Bob", "Carol","Danial"], 400)
add_expense("Danial", ["Bob", "Carol","Danial"], 200)

def calculate_balance():
    total_paid = expense_matrix.sum(axis=1)
    total_owed = expense_matrix.sum(axis=0)
    balance = total_paid - total_owed
    return balance

def show_results():
    balance = calculate_balance()

    print("\nFinal Balances:")
    for i in range(len(friends)):
        if balance[i] > 0:
            print(f"{friends[i]} will receive ₹{balance[i]:.2f}")
        else:
            print(f"{friends[i]} should pay ₹{-balance[i]:.2f}")

def settle_debts():
    balance = calculate_balance()

    creditors = []
    debtors = []

    for i in range(len(balance)):
        if balance[i] > 0:
            creditors.append((friends[i], balance[i]))
        elif balance[i] < 0:
            debtors.append((friends[i], -balance[i]))

    print("\nSettlement:")
    i, j = 0, 0

    while i < len(debtors) and j < len(creditors):
        debtor, d_amt = debtors[i]
        creditor, c_amt = creditors[j]

        pay = min(d_amt, c_amt)

        print(f"{debtor} pays ₹{pay:.2f} to {creditor}")

        d_amt -= pay
        c_amt -= pay

        debtors[i] = (debtor, d_amt)
        creditors[j] = (creditor, c_amt)

        if d_amt == 0:
            i += 1
        if c_amt == 0:
            j += 1


show_results()
settle_debts()


```
